Binning - Grouping Customers by Age & then using Bar Plot to visualise the distribution

```python
bin = [1,18,31,45,60,100]
bin_labels = ['1-18', '18-30', '31-44', '45-59','60-99'] #pd.cut() is a function provided by the pandas library in Python, which is commonly used for binning or categorizing continuous data into discrete intervals or "bins". It is often used to convert a continuous variable into a categorical one, making it easier to analyze or visualize the data.
bin_counts = pd.cut(data['Age'], bins=bin, right=False).value_counts().sort_index() #right = False means bins are included on left, excluded on right
plt.bar(bin_labels, bin_counts, color = ['blue','maroon','red','purple','cyan'])
plt.xlabel('Age')
plt.ylabel('Count of person')
plt.title("Distribution of Customers across Age groups")
plt.show()
```

```python
print(bin_counts)
```

We convert Categorical variables : Gender & satisfaction into Numerical value sets using one-hot encoding

```python
data_encoded = pd.get_dummies(data, columns=['Gender', 'satisfaction']) #The pd.get_dummies function is used to perform one-hot encoding on the specified columns. It will create new columns for each category in the categorical attributes with binary values (0 or 1).
```

```python
data_encoded.head()
```

```python
data['satisfaction'].value_counts()
```

We sum up all the Customer Ratings into one single column called 'Customer Ratings'

```python
df = data_encoded.iloc[:,5:14]
data_encoded['Customer_Ratings'] = df.sum(axis=1) # axis = 1 considers all the values for every column
```

```python
data_encoded.head()
```

We use Covariance Matrix in order to explore relationships between the two numerical variables : Age & Customer_Ratings

```python
age_ratings_data = data_encoded[['Age', 'Customer_Ratings']]
covariance_matrix = np.cov(age_ratings_data,rowvar=False) #rowvar is set to False (the default), it means that each column of the input array represents a variable, and each row represents an observation.
covariance_matrix
```

Covariance Matrix Heatmap visualised for Age & Customer_Ratings:

```python
plt.figure(figsize=(8, 6))
plt.title('Covariance Matrix Heatmap representation between variables : Age & Customer_Ratings')
cov_heatmap = plt.imshow(covariance_matrix, cmap='coolwarm', interpolation='nearest') #interpolation='nearest', it means that the heatmap will display the colors in a way that the transition between adjacent cells is abrupt and doesn't involve any smoothing or blending.

# Displaying values in each cell
for i in range(covariance_matrix.shape[0]):
    for j in range(covariance_matrix.shape[1]):
        plt.text(j, i, f'{covariance_matrix[i, j]:.2f}', ha='center', va='center', color='white')

plt.xticks(np.arange(len(age_ratings_data.columns)), age_ratings_data.columns,)
plt.yticks(np.arange(len(age_ratings_data.columns)), age_ratings_data.columns)
plt.colorbar(cov_heatmap)

plt.show()
```

Chi-Square Test of Independence performed on two categorical variables : Gender & satisfaction

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(data['Gender'], data['satisfaction'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-Square:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```

Test Result : P-value less than 0.05 signifies no relationship among the  variables : Gender and Satisfaction, and thus we fail to reject Null Hypothesis

Stacked Bar Plot used to visualise Chi-Sqaure Result

```python
contingency_table.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Stacked Bar Plot: Gender vs. Satisfaction')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Satisfaction')
plt.show()
```

Comparing Customer Ratings for satisfied and dissatisfied customers using Independent T-Test

```python
from scipy.stats import ttest_ind
satisfied_ratings = data_encoded[data_encoded['satisfaction_satisfied'] == 1]['Customer_Ratings']
dissatisfied_ratings = data_encoded[data_encoded['satisfaction_dissatisfied'] == 1]['Customer_Ratings']
t_statistic, p_value = ttest_ind(satisfied_ratings, dissatisfied_ratings)
print("Independent T-Test Results:")
print("t-statistic:", t_statistic)
print("p-value:", p_value)
# Inference based on the p-value
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Test Result : Reject the null hypothesis. There is a significant difference between customer ratings.")
else:
    print("Test Result : Fail to reject the null hypothesis. There is no significant difference between customer ratings.")
```

Finally, we use boxplot to visualise t-test results for the two groups - satisfied & dissatisfied customers on Customer Ratings

```python
plt.figure(figsize=(8, 6))
box_data = [satisfied_ratings, dissatisfied_ratings]
plt.boxplot(box_data, labels=['Satisfied', 'Dissatisfied'])
plt.title('Box Plot: Customer Rating by Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Customer Rating')
plt.show()
```
